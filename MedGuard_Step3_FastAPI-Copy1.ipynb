```python
# Cell 5 — Diabetes prediction test
import requests, json

patient = {
    "Pregnancies"            : 6,
    "Glucose"                : 148,
    "BloodPressure"          : 72,
    "SkinThickness"          : 35,
    "Insulin"                : 0,
    "BMI"                    : 33.6,
    "DiabetesPedigreeFunction": 0.627,
    "Age"                    : 50
}

r = requests.post('http://localhost:8000/predict/diabetes', json=patient)
result = r.json()

print('=' * 50)
print('  DIABETES PREDICTION RESULT')
print('=' * 50)
print(f"  Risk Score  : {result['probability']}%")
print(f"  Prediction  : {result['prediction']}")
print(f"  Risk Zone   : {result['risk_zone']}")
print()
print('  Top Risk Factors:')
for f in result['top_factors']:
    arrow = '↑' if f['impact'] == 'increases_risk' else '↓'
    print(f"    {arrow} {f['feature']:30s} {f['shap_value']:+.4f}")
```

```python
# Cell 7 — Combined risk score test
import requests, json

combined_data = {
    "diabetes": {
        "Pregnancies"            : 6,
        "Glucose"                : 148,
        "BloodPressure"          : 72,
        "SkinThickness"          : 35,
        "Insulin"                : 0,
        "BMI"                    : 33.6,
        "DiabetesPedigreeFunction": 0.627,
        "Age"                    : 50
    },
    "heart": {
        "HighBP":1,"HighChol":1,"CholCheck":1,"BMI":28.5,
        "Smoker":1,"Stroke":0,"Diabetes":0,"PhysActivity":0,
        "Fruits":0,"Veggies":1,"HvyAlcoholConsump":0,
        "AnyHealthcare":1,"NoDocbcCost":0,"GenHlth":3,
        "MentHlth":5,"PhysHlth":10,"DiffWalk":1,
        "Sex":1,"Age":9,"Education":4,"Income":5
    }
}

r = requests.post('http://localhost:8000/predict/combined', json=combined_data)
result = r.json()

print('=' * 55)
print('  MEDGUARD AI — COMBINED RISK REPORT')
print('=' * 55)
print(f"  Overall Risk Score : {result['risk_score']} / 100")
print(f"  Risk Zone          : {result['risk_zone']}")
print(f"  Alert              : {'YES — HIGH RISK!' if result['alert'] else 'No'}")
print()
print(f"  Diabetes  : {result['diabetes']['probability']}%  — {result['diabetes']['prediction']}")
print(f"  Heart     : {result['heart_disease']['probability']}%  — {result['heart_disease']['prediction']}")
print()
print('  Diabetes Top Factors:')
for f in result['diabetes']['top_factors']:
    arrow = '↑' if f['impact'] == 'increases_risk' else '↓'
    print(f"    {arrow} {f['feature']:30s} {f['shap_value']:+.4f}")
print()
print('  Heart Top Factors:')
for f in result['heart_disease']['top_factors']:
    arrow = '↑' if f['impact'] == 'increases_risk' else '↓'
    print(f"    {arrow} {f['feature']:30s} {f['shap_value']:+.4f}")
```
